```python
import numpy as np
import random as rm

import matplotlib
import matplotlib.pyplot as plt

# The statespace
states = ["Sad","Happy"]

# Possible sequences of events
transitionName = [["HH","HS"],["SS","SH"]]

# Probabilities matrix (transition matrix)
transitionMatrix = [[0.75,0.25],[0.75,0.25]]

        

if sum(transitionMatrix[0])+sum(transitionMatrix[1]) != 2:
    print("Something went wrong. Transition matrix Error")
else: print("Everything is gonna be okay, you should move on!! ;)")
    
    
    
def state_forecast(days):
    # Choose the starting state
#     stateToday = "Happy"
    stateToday = np.random.choice(states)
    print("Start state: " + stateToday)
    # Shall store the sequence of states taken. So, this only has the starting state for now.
    stateList = [stateToday]
    # To calculate the probability of the activityList
    prob = 1
    p1 = 0.75
    p2 = 0.25
    for i in range (0, days):
        if stateToday == "Happy":
            change = np.random.choice(transitionName[0],replace=True,p=transitionMatrix[0])
            if change == "HH":
                prob = prob * p1 
                stateList.append("Happy")
                pass
            elif change == "HS":
                prob = prob * p2
                stateToday = "Sad"
                stateList.append("Sad")
        elif stateToday == "Sad":
            change = np.random.choice(transitionName[1],replace=True,p=transitionMatrix[1])
            if change == "SS":
                prob = prob * p1
                stateList.append("Sad")
                pass
            elif change == "SH":
                prob = prob * p2
                stateToday = "Happy"
                stateList.append("Happy")

#     print("Possible states: " + str(stateList))
#     print("End state after "+ str(days) + " days: " + stateToday)
#     print("Probability of the possible sequence of states: " + str(prob))
    
    # Import LabelEncoder
    from sklearn import preprocessing

    #creating labelEncoder
    le = preprocessing.LabelEncoder()
    state_encoded=le.fit_transform(stateList)
    
    x = stateList
    y = np.arange(0.0, days + 1)

    fig, ax = plt.subplots(figsize=(20, 10))
    ax.plot(y, x)

    ax.set(xlabel='Time (days)', ylabel='State')
    ax.grid()
    plt.show()

# Function that forecasts the possible state for the next 2 days
state_forecast(100)
```
