```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chisquare
from scipy import stats
import statsmodels.api as sm
import pylab

df = pd.read_csv('cervical_cancer.csv')
query = df.loc[:,:'IUD (years)']

age = query['Age']
sex_n = query['Number of sexual partners']
f_sex = query['First sexual intercourse'] 
n_preg = query['Num of pregnancies']  
smokes = query['Smokes'] 
smokes_y = query['Smokes (years)'] 
smokes_py = query['Smokes (packs/year)'] 
h_contra = query['Hormonal Contraceptives'] 
h_contra_y = query['Hormonal Contraceptives (years)'] 
iud = query['IUD']  
iud_y = query['IUD (years)']

def rep_quest(col):
    for n, i in enumerate(col):
        if i == "?":
            col[n] = 0
    col = pd.to_numeric(col)
    return col
def calc_stats(col):
    col_mean = np.mean(col)
    col_median = np.median(col)
    col_mode = stats.mode(col)
    col_max = np.max(col)
    col_min = np.min(col)
    anderson = stats.anderson(col, dist='norm')
    shapiro = stats.shapiro(col)
    print('\nShapiro-Wilk test for a normal distribution: ',shapiro)
    print('\nAnderson test for a normal distribution: ',anderson)
    print('\n',stats.kstest(col, 'norm', args=(np.mean(col), np.std(col) )))
    sm.qqplot(col, loc = np.mean(col), scale = 3, line='s')
    pylab.show()
#     result = "\nMean: " + col_mean + "'\nMedian: " + col_median + "\nMax: " + col_max + "\nMin: " + col_min 
    result = f"\nMean: {col_mean}".format(col_mean) + f"\nMedian: {col_median}".format(col_median) + f"\nMode: {col_median}".format(col_mode) + f"\nMax: {col_max}".format(col_max) + f"\nMin: {col_min}".format(col_min)
    return result

age = rep_quest(age)
sex_n = rep_quest(sex_n)
f_sex = rep_quest(f_sex)
n_preg = rep_quest(n_preg)

print("\nAge ",calc_stats(age))
print("\nNumber of sexual partners ",calc_stats(f_sex))
print("\nFirst sexual intercourse ",calc_stats(f_sex))
print("\nNum of pregnancies ",calc_stats(n_preg))



query['Number of sexual partners'] = rep_quest(age)
query['Number of sexual partners'] = rep_quest(sex_n)
query['First sexual intercourse'] = rep_quest(f_sex) 
query['Num of pregnancies'] = rep_quest(n_preg) 
query['Smokes'] = rep_quest(smokes)
query['Smokes (years)'] = rep_quest(smokes_y)
query['Smokes (packs/year)'] = rep_quest(smokes_py)
query['Hormonal Contraceptives'] = rep_quest(h_contra)
query['Hormonal Contraceptives (years)'] = rep_quest(h_contra_y)
query['IUD']  = rep_quest(iud)
query['IUD (years)'] = rep_quest(iud_y)

# def test(col):
#     anderson = stats.anderson(col, dist='norm')
#     shapiro = stats.shapiro(col)
#     print('\nShapiro-Wilk test for a normal distribution: ',shapiro)
#     print('\nAnderson test for a normal distribution: ',anderson)
#     print('\n',stats.kstest(col, 'norm', args=(np.mean(col), np.std(col) )))
#     sm.qqplot(col, loc = np.mean(col), scale = 3, line='s')
#     pylab.show()
    
# test(age)

pearson = query.corr(method='pearson')
kendall = query.corr(method='kendall')

print("\nPearson's coefficients of correlation\n",pearson)
print("\nKendall's coefficients of correlation\n",kendall)


```
